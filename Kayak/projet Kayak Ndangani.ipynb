Getting the towns informations (latitude, longitude, country) and storing them in a csv files

```python
##Function that get the coordinates, returns 2 lists: one of towns with coordinates and another if an error occurs
def get_coordinates(town_list):
    with open(f"{mypath}src/towns_location_response.json",'wb') as file:
        towns_location=[]
        towns_location_missing=[]
        for city in town_list:
            response = requests.get(f"{url}{city}&language=fr")
            if response.status_code == 200:
                try:
                    file.write(response.content) ##gross data
                    for element in response.json()['results']:
                        if element['country_code'] =='FR' and element['name']==city:
                            towns_location.append(
                                        {'ville':element['name'],
                                        'latitude':element['latitude'],
                                        'longitude':element['longitude'],
                                        'pays':element['country']}
                            )
                        break
                except:
                    towns_location_missing.append(city)
                    continue
    return towns_location, towns_location_missing

## getting the coordinates
towns_location, towns_location_missing = get_coordinates(pd.read_csv(f'{mypath}src/towns.csv')['ville'].to_list())
```

C. GETTING CITIES METEO DATA FROM THEIR LATITUDE AND LONGITUDE

```python
############ WEATHER DATA FROM THE API
## apparent_temperature_max: Maximum daily apparent temperature
## apparent_temperature_min: minimum daily apparent temperature
## rain_sum: Sum of daily rain
## snowfall_sum: Sum of daily snowfall
## windspeed_10m_max: Maximum wind speed and gusts on a day at 10 meters above ground.
```

```python
## testing the requests
url_forecast = "https://api.open-meteo.com/v1/forecast?"
param_weather = "&daily=apparent_temperature_max,apparent_temperature_min,rain_sum,snowfall_sum,windspeed_10m_max&current_weather=true&timezone=Europe%2FLondon"
link = url_forecast+f"latitude={4.98562}&longitude={4.89554}"+param_weather
response = requests.get(link)

print(response.json())
```

```python
towns_weather_infos = []
towns_weather_infos_missing = []

## function to return a dict
def weather_dic(response,df,indice):
    return {'ville':df['ville'].values[indice],
            'latitude':response.json()['latitude'],
            'longitude':response.json()['longitude'],
            'dates':response.json()['daily']['time'],
            'apparent_temperature_max':response.json()['daily']['apparent_temperature_max'],
            'apparent_temperature_min':response.json()['daily']['apparent_temperature_min'],
            'rain_sum':response.json()['daily']['rain_sum'],
            'snowfall_sum':response.json()['daily']['snowfall_sum'],
            'windspeed_10m_max':response.json()['daily']['windspeed_10m_max']}


## Function returning weather informations and the towns, having as input 'latitude' and 'longitude' of the location dataframe
def weather_func(df):
    towns_weather_infos = []
    towns_weather_infos_missing = []
    url_forecast = "https://api.open-meteo.com/v1/forecast?"
    param_weather = "&daily=apparent_temperature_max,apparent_temperature_min,rain_sum,snowfall_sum,windspeed_10m_max&current_weather=true&timezone=Europe%2FLondon"
    list_links = [url_forecast+f"latitude={latitude}&longitude={longitude}"+param_weather \
                for latitude,longitude in zip(df['latitude'].values,df['longitude'].values)]
    for indice,link in enumerate(list_links):
        response = requests.get(link)
        if response.status_code == 200:
            try:
                towns_weather_infos.append(weather_dic(response,df,indice))
            except:
                towns_weather_infos_missing.append(df['ville'].values[indice])
                continue
        else:
            towns_weather_infos_missing.append(df['ville'].values[indice])
    return towns_weather_infos, towns_weather_infos_missing
```

Sorting the dataframe by the weather criteria defined by the function bellow

```python
## function that return a citeria to sort the dataframe with the the weather criteria
## 2 seasons are defined
def current_period(df):
    ## A nice weather for me depends on the current period.
    ## From march to october a nice weather is defined by max 'apparent_temperature_max','apparent_temperature_min'
    ## and min 'rain_sum','snowfall_sum','windspeed_10m_max'.
    sun_by = ['dates','apparent_temperature_max','apparent_temperature_min','rain_sum','snowfall_sum','windspeed_10m_max']
    sun_sort = [True,False,False,True,True,True]

    ## From november to february a nice weather is defined by max 'snowfall_sum' and min 'apparent_temperature_max',
    ## 'apparent_temperature_min','rain_sum','windspeed_10m_max'
    snow_by = ['dates','snowfall_sum','apparent_temperature_max','apparent_temperature_min','rain_sum','windspeed_10m_max']
    snow_sort = [True,False,True,True,True,True]

    ## the 2 seasons
    period = datetime.now().date()
    sunny_period = period >= datetime.strptime('01-03-2022',"%d-%m-%Y").date() and period < datetime.strptime('01-11-2022',"%d-%m-%Y").date()
    snow_period = period >= datetime.strptime('01-11-2022',"%d-%m-%Y").date() and period < datetime.strptime('01-03-2022',"%d-%m-%Y").date()

    if sunny_period:
        criteria = [sun_by,sun_sort]
    else:
        criteria = [snow_by,snow_sort]

    return criteria
```

```python
## The 5 best destinations per day, return a dataset of the the 5 best destinations per day
def get_best_destinations(df):
    liste=[]
    for date in df['dates'].unique():
        ## getting a list of 7 dataframes (one per date), each dataframe having 5 best destinations (5 rows)
        df_temp = df[df['dates']==date].reset_index().drop('index',axis=1).sort_values(by=sorting[0],ascending=sorting[1])
        df_temp=df_temp.loc[:,['ville','latitude','longitude',*sorting[0]]].drop_duplicates().reset_index().drop('index',axis=1).head()
        liste.append(df_temp)
    return liste

liste = get_best_destinations(df_towns_enriched)
df_destinations = pd.concat([destination for destination in liste],ignore_index=True)
df_destinations.to_csv(f'{mypath}src/best_destinations.csv',index=None)
display(df_destinations.head())
df_destinations.info()
```
