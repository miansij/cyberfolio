ON A BESOIN D'ANALYSER LES HEURES, MOIS, JOURS, JOURS DE SEMAINE, MINUTES. ON VA DONC CREER CES FEATURES A PARTIR DE LA DATE DONT LE TYPE VA ETE TANSFORMEE EN DATETIME GRACE A LA FONCTION SUIVANTE.

```python
## fonction pour transformer 'Date/Time' en datetime et créer des colonnes dont on aurait besoin idéalement:
## 'year,month,dayofweek,day,hour,minute'
def func_change_type_date(df):
    df1=df
    df1['date']=pd.to_datetime(df1['Date/Time'])
    df1['year']=df1['date'].dt.year
    df1['month']=df1['date'].dt.month
    df1['dayofweek']=df1['date'].dt.dayofweek
    df1['day']=df1['date'].dt.day
    df1['hour']=df1['date'].dt.hour
    df1['minute']=df1['date'].dt.minute
    df1=df1.drop('Date/Time',axis=1)
    return df1
```

Vu le grand nombre de données et pour ne pas ralentir le travail d'exploration et de visualisation, nous utiliserons un dataset constitué d'un échantillon.

```python
## fonction qui prend des dataframes, les concatene et en renvoie un sample, appelle la fonction qui transforme la date en datetime
## crée des features et retourne un dataframe allégé
def func_df_samples(*dfs, sample_size=None,frac_size=None):
  df=pd.concat([*dfs]) ## concaténation des samples
  if sample_size:
    df = df.sample(n=sample_size,random_state=7)
  elif frac_size:
    df = df.sample(frac=frac_size,random_state=7)
  else:
    df = df.sample(random_state=7)
  ## appel à une fonction pour transformer le type des dates et crée des features
  df = func_change_type_date(df)
  print('feature engineering ok')
  return df
```

```python
## On appelle la fonction
df_sample = func_df_samples(*liste_df,sample_size=50000)
```

Preprocessing avec scikit-learn

On va essayer KMeans et DBSCAN avec un petit nombre de cluster pour voir lequel donne un meilleur découpage de la zone

```python
def func_scatter_map(df,Lat,Lon,title='pickup zones',animation_frame=None,size=None,zoom=8):
  if animation_frame==None:
    animation_frame=None
  if size==None:
    size_=None

  mapbox_style=["open-street-map", "carto-positron", "carto-darkmatter", "stamen-terrain", "stamen-toner","stamen-watercolor"]
  fig = px.scatter_mapbox(df,
                      lat='Lat',
                      lon='Lon',
                      color= 'cluster',
                      opacity=0.8,
                      title= title,
                      animation_frame=animation_frame,
                      size=size,
                      height=800,
                      zoom=zoom,
                      mapbox_style=mapbox_style[0],
                      center={'lat':df['Lat'].median(),'lon':df['Lon'].median()},
                      )

  fig.show()
```

Fonction pour lancer le clustering.

```python
## Fonction pour lancer le clustering sous KMeans ou DBSCAN
def func_pickups(dataframe,model,drop_col=None):#,param1=None,param2=None,param3=None):
  X=dataframe.drop(drop_col,axis=1)

  scaler = StandardScaler()
  num_vars = ['Lat','Lon']
  X_transf = scaler.fit_transform(X[num_vars])
  labels=model.fit_predict(X_transf)
  print("le nombre de clusters")
  print(len(np.unique(labels)))
  ## on crée la colonne des clusters
  X['cluster']=labels

  return X
```

Test sur Kmeans

```python
X_kmeans=func_pickups(df,KMeans(n_clusters=24, random_state=1, n_init=1),col_to_drop)
func_scatter_map(X_kmeans,'Lat','Lon')
```

```python
X_dbscan=func_pickups(df,DBSCAN(eps=0.1, min_samples=6, metric="manhattan"),col_to_drop)
func_scatter_map(X_dbscan,'Lat','Lon')
```

Cette baseline montre que avec DBSCAN on trouve certains clusters très grands et beaucoup de très petits. Les clusters avec Kmeans sont un peu mieux équilibrés, ce qui donne un meilleur découpage.

RECOMMANDATION DE HOTS ZONES

```python
## Fonction de recommendation de HOT ZONES, en fonction du mois et du jour de semaine choisis. Les heures seront choisis à partir de la map.
## La taille des points montre l'importance de la zone.
def recomandation(df,month,dayofweek,title=None,animation_frame=None,size=None,zoom=None):
  X = df
  labels = X['cluster']
  ## on crée le point central de chaque cluster
  dic_cluster_median={}
  for x in np.unique(labels):
    dic_cluster_median[x]=[X[X['cluster']==x]['Lat'].median(),X[X['cluster']==x]['Lon'].median()]
  X['Lat_med']=[dic_cluster_median[i][0] for i in labels]
  X['Lon_med']=[dic_cluster_median[i][1] for i in labels]
  X['size']=[X[X['cluster']==x]['cluster'].count() for x in labels]

  ## les jours de semaine
  X['joursemaine']=X['dayofweek'].map({0:"dimanche",1:"lundi",2:"mardi",3:"mercredi",4:"jeudi",5:"vendredi",6:"samedi"})

  X=X.sort_values('hour',ascending=True)[(X['month']==month)&(X['dayofweek']==dayofweek)]
  X.drop(['Lat','Lon'],axis=1,inplace=True)
  X.rename(columns ={'Lat_med':'Lat','Lon_med':'Lon'},inplace=True)

  func_scatter_map(X,'Lat','Lon',animation_frame=animation_frame,title= title,size=size,zoom=zoom)
  return X
```

```python
## recommendation de hot zones pour le mois 9 et le jour de semaine 4. Cet exemple ne porte que sur le mois de septembre.
X_recom=recomandation(X_pick,9,4,animation_frame='hour',title= f'UBER HOT ZONES',size='cluster',zoom=10)
```
